```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 데이터 세트 만들기
n_data = 1000
data_set = []

# 목표 함수: y = 0.5 * x + 0.2 (W = 0.5, b = 0.2)
for i in range(n_data):
    _x = np.random.normal(0.0, 0.1)
    _y = 0.5 * _x + 0.2 + np.random.normal(0.0, 0.02)
    data_set.append([_x, _y])
    
x = [v[0] for v in data_set]
y = [v[1] for v in data_set]

print(np.array(x).shape)
print(np.array(y).shape)

plt.plot(x, y, 'ro')
plt.show()
```

```python
# set Variable(weight, bias)
W = tf.Variable(np.random.normal([1]))
b = tf.Variable(np.random.normal([1]))

# model
_y = W * x + b
```

```python
# Func
loss = tf.reduce_mean(tf.square(_y - y)) # Mean Squared Error(MSE)
optimizer = tf.train.GradientDescentOptimizer(0.5) # set optimizer and learning rate
train = optimizer.minimize(loss)
```

```python
# Open Session and Initializing Variable
sess = tf.Session()
init = tf.global_variables_initializer()
sess.run(init)

# 50회 학습
for step in range(50):
    sess.run(train)

print(sess.run(W), sess.run(b))
plt.plot(x, y, 'ro')
plt.plot(x, sess.run(W) * x + sess.run(b))
plt.show()
```

```python
init = tf.global_variables_initializer()
sess.run(init)

# 200회 학습
for step in range(200):
    sess.run(train)

print(sess.run(W), sess.run(b))
plt.plot(x, y, 'ro')
plt.plot(x, sess.run(W) * x + sess.run(b))
plt.show()
```
