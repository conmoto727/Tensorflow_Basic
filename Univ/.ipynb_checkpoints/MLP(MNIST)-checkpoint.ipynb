```python
import numpy as np
import os
import tensorflow as tf
from scipy.misc import imread, imshow
from tensorflow.examples.tutorials.mnist import input_data
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
cwd = os.getcwd()
print("Current directory is %s" % (cwd))
# img info def
def print_info(img):
    print("Type: %s" %(type(img)))
    print("Shape: %s" %(img.shape,))
```

```python
mnist = input_data.read_data_sets('data/', one_hot=True)
```

```python
_input = 784
_hidden_1 = 256
_hidden_2 = 128
_classes = 10

x = tf.placeholder("float", [None, _input])
y = tf.placeholder("float", [None, _classes])

stddev = 0.1
weights ={
    'h1': tf.Variable(tf.random_normal([_input, _hidden_1], stddev=stddev)),
    'h2': tf.Variable(tf.random_normal([_hidden_1, _hidden_2], stddev=stddev)),
    'out': tf.Variable(tf.random_normal([_hidden_2, _classes], stddev=stddev))
}
biases = {
    'b1': tf.Variable(tf.random_normal([_hidden_1])),
    'b2': tf.Variable(tf.random_normal([_hidden_2])),
    'out': tf.Variable(tf.random_normal([_classes]))
}
```

```python
def MLP(_x, _weight, _biases):
    layer_1 = tf.nn.sigmoid(tf.add(tf.matmul(_x, _weight['h1']),_biases['b1']))
    layer_2 = tf.nn.sigmoid(tf.add(tf.matmul(layer_1, _weight['h2']),_biases['b2']))
    return (tf.matmul(layer_2, _weight['out']) + _biases['out'])

pred = MLP(x, weights, biases)

cost = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(labels=y, logits=pred))
optm = tf.train.AdamOptimizer(learning_rate=0.001).minimize(cost)
corr = tf.equal(tf.argmax(pred, 1), tf.argmax(y, 1))
accr = tf.reduce_mean(tf.cast(corr, "float"))

init = tf.global_variables_initializer()
```

```python
epochs = 12
batch_size = 100
step = 3

sess = tf.Session()
sess.run(init)

for epoch in range(epochs):
    avg_cost = 0.
    total_batch = int(mnist.train.num_examples/batch_size)
    
    for i in range(total_batch):
        batch_xs, batch_ys = mnist.train.next_batch(batch_size)
        sess.run(optm, feed_dict = {x: batch_xs, y: batch_ys})
        avg_cost += sess.run(cost, feed_dict = {x: batch_xs, y: batch_ys})
    avg_cost = avg_cost/total_batch
    
    if(epoch+1) % step == 0:
        print("Epoch: %03d/%03d cost: %.9f" %(epoch, epochs, avg_cost))
        
        Train_set = {x: batch_xs, y: batch_ys}
        train_acc = sess.run(accr, feed_dict=Train_set)
        print ("TRAIN ACCURACY: %.3f" % (train_acc))
        
        Test_set = {x: mnist.test.images, y: mnist.test.labels}
        test_acc = sess.run(accr, feed_dict=Test_set)
        print ("TEST ACCURACY: %.3f" % (test_acc))
```

```python
testimg   = mnist.test.images
testlabel = mnist.test.labels
print(testimg.shape)

#test 5 images
nsample = 5
randidx = np.random.randint(testimg.shape[0], size = nsample)
print(randidx)

predict = tf.argmax(pred,1)
for i in randidx:
    curr_img = np.reshape(testimg[i,:], (28,28))
    curr_pred = sess.run(predict, feed_dict={x:testimg[i:i+1,:]})
    curr_label = np.argmax(testlabel[i,:])
    plt.matshow(curr_img, cmap = plt.get_cmap('gray'))
    plt.title(""+ str(i) + "th Test Data Label is "+str(curr_label))
    plt.show()
    print("%dth Test Data predict num: %d"%(i, curr_pred))
```
