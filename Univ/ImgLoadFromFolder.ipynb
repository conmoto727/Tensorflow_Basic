```python
import matplotlib.pyplot as plt
import numpy as np
import os
from scipy.misc import imread, imresize
%matplotlib inline
```

```python
cwd = os.getcwd()
path = cwd + "/Cats"
valid = [".jpg", ".jpeg"]
print("%s Files in %s:\n%s" %(len(os.listdir(path)), path, os.listdir(path)))
```

```python
for f in os.listdir(path):
    name = os.path.splitext(f)[0]
    print(name)
    
for f in os.listdir(path):
    exe = os.path.splitext(f)[1]
    print(exe)
```

```python
imgs = []
names = []
for f in os.listdir(path):
    ext = os.path.splitext(f)[1]
    if ext.lower() not in valid:
        continue
    fullpath= os.path.join(path, f)
    
    imgs.append(imread(fullpath)) #이미지 또한 리스트의 요소에 imread로 저장할 수 있다.
    names.append(os.path.splitext(f)[0]+os.path.splitext(f)[1])
    
print("%d images loaded" %(len(imgs)))
```

```python
idx = np.sort(np.random.randint(30, size = 5))
print(idx)
```

```python
nimgs = len(imgs)
randidx = np.random.randint(nimgs, size=3)
for i in randidx:
    curr_img = imgs[i]
    curr_name = names[i]
    plt.figure(i)
    plt.imshow(curr_img)
    plt.title(curr_name)
    plt.draw()
```
