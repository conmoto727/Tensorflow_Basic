```python
import numpy as np
import os
from scipy.misc import imread, imresize
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# load current directory
cwd = os.getcwd()
print("%s" % (cwd))
```

```python
# img info def
def print_info(img):
    print("Type: %s" %(type(img)))
    print("Shape: %s" %(img.shape,))
```

```python
# Load
Cat = imread(cwd + "/Cat.jpg")
print_info(Cat)
```

```python
plt.figure(0)
plt.imshow(Cat)
plt.title("Image with imread")
plt.draw()
```

```python
def rgb2gray(rgb):
    if len(rgb.shape) is 3: #3차원 이미지(rgb이미지)
        return np.dot(rgb[...,:3], [0.2, 0.5, 0.1]) #rgb 차원을 곱해서 더하면서 차원 하나 줄어든다.
    else:
        print("Current Image if Gray!")
        return rgb

Test_a = np.array([[[1,2,3],[3,4,5]],[[5,6,6],[7,8,7]]])
print(Test_a.shape)
print(len(Test_a.shape))
Test_b = np.array([[1,2,3,4],[5,6,7,8]])
print(Test_b.shape)
print(len(Test_b.shape))

# len(.shape)은 차원의 길이를 반환!!!
```
